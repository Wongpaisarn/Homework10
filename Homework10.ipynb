<h1 style="text-align: center">
    Lucus-Kanade Sparse Optical Flow 
</h1> <br>
<h2 style="text-align: left">
    <a href="https://docs.opencv.org/4.5.3/dc/d6b/group__video__track.html#ga473e4b886d0bcc6b65831eb88ed93323"> calcOpticalFlowPyrLK()</a>
    is a Lucas-Kanade Spares Optical Flow built-in OpenCV function inside 
    <a href="https://docs.opencv.org/4.5.3/dc/d6b/group__video__track.html"> Object Tracking </a>
    module.
</h2>

```python
def lucas_kanade_optical_flow(video_device) :

    cap = cv2.VideoCapture(video_device)
    # params for ShiTomasi corner detection
    feature_params = dict( maxCorners = 500,
                        qualityLevel = 0.03,
                        minDistance = 7,
                        blockSize = 25 )

    # Parameters for lucas kanade optical flow
    lk_params = dict( winSize  = (21,21),
                    maxLevel = 3,
                    criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

    #Create some random colors
    colors = np.random.randint(0, 255, (500, 3)) # 500 values 3 channel

    #Take first frame and find corner
    ret, old_frame = cap.read()
    old_gray = cv2.cvtColor(old_frame, cv2.COLOR_BGR2GRAY)
    p0 = cv2.goodFeaturesToTrack(old_gray, mask=None, **feature_params) # Feature detection, Harris corner with Shi-Tomasi response function

    # Create a mask image for drawing overlay
    mask = np.zeros_like(old_frame)

    while cap.isOpened() :
        
        ret, frame = cap.read()

        if ret :
            frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            #calculate optical flow 
            p1, st, err = cv2.calcOpticalFlowPyrLK(
                old_gray, frame_gray, p0, None, **lk_params
            )

            # Select good points
            good_new = p1[st == 1]
            good_old = p0[st == 1]

            # Traceline drawing
            for i, (new, old) in enumerate(zip(good_new, good_old)):
                a, b = new.ravel().astype(int)
                c, d = old.ravel().astype(int)
                mask = cv2.line(mask, (a, b), (c, d), colors[i].tolist(), 2)
                frame = cv2.circle(frame, (a,b), 5, colors[i].tolist(), -1)
            
            compare_img = cv2.hconcat([frame, mask])
            disp_img = cv2.add(frame, mask)
            cv2.imshow('compare', compare_img)
            cv2.imshow('frame', disp_img)

            key = cv2.waitKey(27) & 0xFF
            if key == 27 or key == ord('q') :
                break
            elif key == ord('c') : # clear mask
                mask = np.zeros_like(old_frame)
                p0 = cv2.goodFeaturesToTrack(old_gray, mask=None, **feature_params)
            else :
                #Update the previous frame and previous points
                old_gray = frame_gray.copy()
                p0 = good_new.reshape(-1, 1, 2)
        else :
            break
    # จบลูป while
    cap.release()
    # แสดง video โดยใช้หน้าต่าง
    cv2.destroyAllWindows()

lucas_kanade_optical_flow(0)
```

<h1 style="text-align: center">
    Farnebäck Optical Flow
</h1> <br>
<h2 style="text-align: left">
    <a href="https://docs.opencv.org/4.5.3/dc/d6b/group__video__track.html#ga5d10ebbd59fe09c5f650289ec0ece5af"> calcOpticalFlowFarneback()</a>
    is a Farnebäck Dense Optical Flow built-in OpenCV function inside 
    <a href="https://docs.opencv.org/4.5.3/dc/d6b/group__video__track.html"> Object Tracking </a>
    module. <br>
    <a href="https://docs.opencv.org/4.5.3/d2/de8/group__core__array.html#gac5f92f48ec32cacf5275969c33ee837d">cartToPolar() </a>
    is a utility function for cartesian to polar conversion
</h2>

```python
def farneback_dense_optical_flow(video_device) :
    cap = cv2.VideoCapture(video_device)

    ret, frame = cap.read()
    last_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    hsv = np.zeros_like(frame)
    hsv[:, :, 1] = 255

    while cap.isOpened() :
        ret, frame = cap.read()

        if ret :
            
            frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            flow = cv2.calcOpticalFlowFarneback(last_frame, frame_gray, None, 0.5, 3, 15, 3, 5, 1.2, 0) # flow dx dy
            
            mag, ang = cv2.cartToPolar(flow[:, :, 0], flow[:, :, 1])
            hsv[:, :, 0] = ang*(180/np.pi/2)
            hsv[:, :, 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)

            flow_rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

            # thresh = cv2.inRange(hsv,(0, 0, 30), (20,255,255))
            # motion_segment = cv2.bitwise_and(frame, frame, mask=thresh )
            # cv2.imshow('thresh', motion_segment)
           
            last_frame = frame_gray.copy()

            cv2.imshow('frame', frame)
            cv2.imshow('flow', flow_rgb)
            key = cv2.waitKey(27) & 0xFF
            if key == 27 or key == ord('q') :
                break

        else :
            break
    cap.release()
    cv2.destroyAllWindows()
farneback_dense_optical_flow('./videos/Flow Visualization.mp4')
```

```python
# เปิดฟังก์ชั่นเพื่อทำการเปิด video grandcentral.mp4
def lucas_kanade_optical_flow(video_device) :
    # ฟังก์ชันที่ใช้ในการอ่านไฟล์วิดีโอจากข้อมูลในคอมพิวเตอร์เก็บใส่ในตัวแปร video_device
    cap = cv2.VideoCapture(video_device)
    # นำข้อมูลทั้งหมดที่อยู่ใน dict เก็บไว้ในตัวแปล feature_params
    feature_params = dict( maxCorners = 500, 
                        qualityLevel = 0.03,
                        minDistance = 7,
                        blockSize = 25 )
    # นำข้อมูลทั้งหมดที่อยู่ใน dict เก็บไว้ในตัวแปล lk_params
    lk_params = dict( winSize  = (21,21),
                    maxLevel = 3,
                    criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))
    # ทำการสร้างสีสุ่ม
    colors = np.random.randint(0, 255, (500, 3)) 
    # อ่าน cap.read() โดย ret จะเก็บค่า true กับ false และ old_frame จะเก็บภาพไว้
    ret, old_frame = cap.read()
    # ทำการเปลี่ยน old_frame ให้เป็น gray scale แล้วเก็บไว้ในตัวแปร old_gray
    old_gray = cv2.cvtColor(old_frame, cv2.COLOR_BGR2GRAY)
    # ทำการกำหนดจุดของวัตถุที่เราต้องการทำการติดตาม แล้วเก็บไว้ในตัวแปร p0
    p0 = cv2.goodFeaturesToTrack(old_gray, mask=None, **feature_params) 
    # ทำการ mask ที่ภาพ old_frame
    mask = np.zeros_like(old_frame)
    # กำหนดจุดขอบเขตที่ต้องการ แล้วเปลี่ยนให้เป็น int 32 bit
    pts = np.array([[172,33], [33,480], [720,480], [560,30]], np.int32)
    # วาดภาพพื้นหลังให้เป็นสีดำ
    detec = np.zeros(old_frame.shape[:2], np.uint8)
    # วาดภาพจุดที่เราสนใจให้เป็นสีขาว
    cv2.drawContours(detec,[pts],-1,(255,255,255),-1,cv2.LINE_AA)
    # เมื่อเปิด video จะเข้าลูป while
    while cap.isOpened() :
    # อ่าน cap.read() โดย ret จะเก็บค่า true กับ false และ frame จะเก็บภาพไว้
        ret, frame = cap.read()
        # ถ้า ret เป็น true
        if ret :
            # ทำการเปลี่ยน frame ให้เป็น gray scale แล้วเก็บไว้ในตัวแปร frame_gray
            frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            # คำนวณหาการเคลื่อนที่ของวัตถุเพื่อทำการติดตาม
            p1, st, err = cv2.calcOpticalFlowPyrLK(old_gray, frame_gray, p0, None, **lk_params)
            # เลือกจุดที่ดี
            good_new = p1[st == 1]
            good_old = p0[st == 1]
            # เลือกจุดที่ดีทั้งหมดใน video
            for i, (new, old) in enumerate(zip(good_new, good_old)):
                a, b = new.ravel().astype(int) #เปลี่ยน ravel ให้เป็น int
                c, d = old.ravel().astype(int) #เปลี่ยน ravel ให้เป็น int
                # ถ้า b >= detec.shape[0] ให้ b = detec.shape[0]-1
                if (b >= detec.shape[0]):
                    b = detec.shape[0]-1
                # ถ้า a >= detec.shape[0] ให้ a = detec.shape[0]-1   
                if (a >= detec.shape[1]):
                    a = detec.shape[1]-1
                # ถ้า detec[b,a] == 255
                if detec[b,a] == 255:
                   # ทำการเขียนเส้นสีแดงตามจุดที่สนใจ
                   mask = cv2.line(mask, (a, b), (c, d), (0,0,255), 1)
                   # ทำการวาดวงกลมตามจุดที่สนใจ
                   frame = cv2.circle(frame, (a,b), 5, (0,255,0), 1)
            # ทำการเขียนเส้นตามวงกลมเก็บไว้ในตัวแปร compare_img
            compare_img = cv2.hconcat([frame, mask])
            # นำ frame มารวมกับ mask
            disp_img = cv2.add(frame, mask)
            # วาดกรอบจุดที่ได้กำหนดไว้ในตัวแปร pts ใน disp_img
            cv2.drawContours(disp_img,[pts],-1,(255,0,0),1,cv2.LINE_AA)
            # แสดง ompare_img กับ disp_img
            cv2.imshow('compare', compare_img)
            cv2.imshow('frame', disp_img)
            key = cv2.waitKey(33) & 0xFF
            # ถ้า กดปุ่ม esc กับ q จะทำการปิดหน้าต่าง video
            if key == 27 or key == ord('q') :
                break
            # good_new < 480 จะทำการกำหนดจุดของวัตถุที่เราต้องการทำการติดตาม แล้วเก็บไว้ในตัวแปร p0 ใหม่
            elif len(good_new) <480 : 
                p0 = cv2.goodFeaturesToTrack(old_gray, mask=None, **feature_params)
            # อัปเดตเฟรมก่อนหน้าและจุดก่อนหน้า
            else :
                old_gray = frame_gray.copy()
                p0 = good_new.reshape(-1, 1, 2)
       # เมือ false หรือ video เล่นจบจะการทำปิดหน้าต่าง video
        else :
            break
    # c จบลูป while
    cap.release()
    # แสดง video โดยใช้หน้าต่าง
    cv2.destroyAllWindows()
# นำ video เข้าฟังก์ชั่น farneback_dense_optical_flow
lucas_kanade_optical_flow ('./videos/grandcentral.mp4')
```
